```python
from fastapi import FastAPI, HTTPException
import requests
import nest_asyncio
import uvicorn

nest_asyncio.apply()

# Inisialisasi FastAPI
app = FastAPI()

# API key untuk Fixer.io
API_KEY = "0c06f2764bfc96cb92dfd1eff60ca0be"
BASE_URL = "http://data.fixer.io/api/latest"

@app.get("/convert_rate/")
def get_convert(base_currency: str, target_currency: str):
    """
    Mengambil nilai tukar mata uang berdasarkan mata uang asal dan tujuan.
    """
    # Mengatur parameter API
    params = {
        "access_key": API_KEY,  # API key untuk autentikasi
        "symbols": f"{base_currency},{target_currency}"  # Simbol mata uang yang dibutuhkan
    }

    try:
        # Memanggil API
        response = requests.get(BASE_URL, params=params)
        if response.status_code != 200:
            raise HTTPException(status_code=response.status_code, detail="Error fetching conversion rate data.")
        
        data = response.json()

        # Mengecek apakah data valid dan memiliki key 'rates'
        if "rates" not in data:
            raise HTTPException(status_code=500, detail="Error in API response structure.")
        
        # Ambil kurs dari EUR ke mata uang dasar dan target
        rates = data["rates"]
        eur_to_base = rates.get(base_currency.upper())
        eur_to_target = rates.get(target_currency.upper())

        # Jika salah satu kurs tidak ditemukan
        if not eur_to_base or not eur_to_target:
            raise HTTPException(status_code=400, detail="Invalid base or target currency.")

        # Hitung kurs dari base_currency ke target_currency
        exchange_rate = eur_to_target / eur_to_base

        return {
            "base_currency": base_currency.upper(),
            "target_currency": target_currency.upper(),
            "exchange_rate": round(exchange_rate, 2)  # Format dengan dua angka desimal
        }

    except requests.exceptions.RequestException as e:
        raise HTTPException(status_code=500, detail=f"Error contacting exchange rate API: {str(e)}")

# Menjalankan server
def run_server():
    uvicorn.run(app, host="127.0.0.1", port=8500)

run_server()
```
